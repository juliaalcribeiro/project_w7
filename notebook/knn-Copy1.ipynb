```python
import pandas as pd
import numpy as np

# separate features and target
y = df["Diagnosis"]
X = df.drop(columns=["Diagnosis"])

# compute IQR bounds per column
Q1 = X.quantile(0.25)
Q3 = X.quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

# boolean mask of outliers (any column outside bounds)
outlier_mask = (X.lt(lower) | X.gt(upper)).any(axis=1)

# indices and count
outlier_indices = df.index[outlier_mask]
n_outliers = outlier_mask.sum()
print(f"Outliers (IQR): {n_outliers} rows")

```

```python
from scipy.stats import zscore

X_z = X.apply(zscore)  # column-wise z-scores
z_thresh = 3.0
outlier_mask_z = (X_z.abs() > z_thresh).any(axis=1)
print(f"Outliers (|z|>{z_thresh}): {outlier_mask_z.sum()} rows")
```

```python
df_clean = df.loc[~outlier_mask].reset_index(drop=True)
df_clean 
```

```python
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
import numpy as np

# choose which dataframe to use: df_clean (dropped), df_wins (capped), or original df
use_df = df  # or df_clean / df_wins

X = use_df.drop(columns=["Diagnosis"]).values
y = use_df["Diagnosis"].values

pipe = Pipeline([
    ("scaler", StandardScaler()),
    ("knn", KNeighborsClassifier(weights="distance"))  # try "uniform" too
])

k_values = [3,5,7,9,11,15,21,31]
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

mean_scores = []
for k in k_values:
    pipe.set_params(knn__n_neighbors=k)
    scores = cross_val_score(pipe, X, y, cv=cv, scoring="accuracy")
    mean_scores.append(scores.mean())

best_idx = int(np.argmax(mean_scores))
best_k = k_values[best_idx]
print("k candidates:", k_values)
print("CV accuracies:", [round(s,4) for s in mean_scores])
print("Best k:", best_k, "CV accuracy:", round(mean_scores[best_idx], 4))
```
